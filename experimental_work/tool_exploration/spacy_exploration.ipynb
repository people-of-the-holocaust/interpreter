```python
import requests
from bs4 import BeautifulSoup
import spacy
```

```python
test_files = ["./Test Files/auflag5_test.txt", "./Test Files/braunschweig_test.txt",
              "./Test Files/dachau_test.txt", "./Test Files/dulagcorinth_test.txt",
              "./Test Files/warsaw_test.txt"]
```

```python
#loading from test data
def get_tokens(filename):
    #loading from test data
    test_file = open(filename, 'r')
    raw_text = test_file.read()
    test_file.close()
    # removing HTML tags
    text = BeautifulSoup(raw_text, "html.parser").get_text(separator=" ", strip=True)
    # load text from article
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    return doc
```

```python
proper_nouns = []
noun_chunks = []
named_entities = []
for file in test_files:
    doc = get_tokens(file)
    propn = []
    chunks = list(doc.noun_chunks)
    ents = list(doc.ents)
    print(f"\n=== Tokens found in {file} ===")
    for token in doc:
        if token.pos_ == "PROPN":
            propn.append(token.text)
        print(token.text, token.pos_, token.dep_)
    proper_nouns.append(propn)
    noun_chunks.append(chunks)
    named_entities.append(ents)
```
