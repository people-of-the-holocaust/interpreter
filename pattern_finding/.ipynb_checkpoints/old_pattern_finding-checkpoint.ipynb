```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# save test file names sorted by volume
test_files = {"v1": [], "v2": [], "v3": [], "v4": []}
v1_docs = [1873, 1778, 1350, 1589, 1368, 2218, 2200, 1391, 1349, 1830]
v2_docs = [3379, 3255, 3195, 2966, 2884, 3423, 2943, 3280, 3042, 2653]
v3_docs = [3960, 3995, 3675, 3582, 4070, 3563, 4243, 3759, 4085, 4088]
v4_docs = [4751, 4624, 4407, 4538, 4828, 4444, 4702, 4896, 4576, 4483]
for i in v1_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v1"].append(fname)
for i in v2_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v2"].append(fname)
for i in v3_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v3"].append(fname)
for i in v4_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v4"].append(fname)
```

```python
# load in people table from scraping process
df = pd.read_csv('../ecg_interpreter/people_table.csv')
df = df.drop('Unnamed: 0', axis=1)
# save names in a list
people_names = df['Name'].tolist()
```

```python
# helper function - find all instances of a person's name, save the result in a dict
def find_names(body):
    # key: person name, value: sentence containing name
    key_sentences = {}
    for name in people_names:
        # if name appears in body, get sentence surrounding the name
        if name in body:
            name_index = body.find(name)
            end_of_sent = body[name_index:].find('.') + 1 + name_index
            start_of_sent = body[:name_index].rfind('.') + 1
            key_sentences[name] = body[start_of_sent:end_of_sent]
    return key_sentences
```

```python
# helper function - print key sentences
def pretty_print(title, dct):
    print("ARTICLE:", title)
    for key, item in dct.items():
        print("NAME:", key)
        print("SENTENCE:", item)
        print()
    print("------------------------------------")
```

```python
# helper function - write key sentences to file
def save_sentences(title, dct):
    with open("people_sentences.txt", "a") as f:
        f.write(f"ARTICLE: {title}\n")
        for key, item in dct.items():
            f.write(f"NAME: {key}\n")
            f.write(f"SENTENCE: {item}\n\n")
        f.write("------------------------------------\n\n")

# helper function - write volume number to file
def new_volume(vol):
    with open("people_sentences.txt", "a") as f:
        f.write("------------------------------------------------------------------------\n")
        f.write(f"VOLUME: {vol}\n\n")
```

```python
for vol, articles in test_files.items():
    new_volume(vol)
    for article_path in articles:
        # convert article html file into soup object
        with open(article_path, "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, 'lxml')
        # get article title and text
        title = soup.title.text
        body_tags = soup.select("div#body")
        if len(body_tags) > 1:
            print("More than 1 body divs:", len(body_tags))
        # should only have one body div, but loop for potential errors
        for body in body_tags:
            text = body.get_text()
            # get sentences with people names - print out & save to file
            key_sentences = find_names(text)
            pretty_print(title, key_sentences)
            save_sentences(title, key_sentences)
```
