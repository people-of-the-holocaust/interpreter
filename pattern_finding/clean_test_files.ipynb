```python
import pandas as pd
from bs4 import BeautifulSoup
import nltk
# nltk.download('punkt') - one time set up
# nltk.download('averaged_perceptron_tagger_eng') - one time set up
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import re
```

```python
# save test file names sorted by volume
test_files = {"v1": [], "v2": [], "v3": [], "v4": []}
v1_docs = [1873, 1778, 1350, 1589, 1368, 2218, 2200, 1391, 1349, 1830]
v2_docs = [3379, 3255, 3195, 2966, 2884, 3423, 2943, 3280, 3042, 2653]
v3_docs = [3960, 3995, 3675, 3582, 4070, 3563, 4243, 3759, 4085, 4088]
v4_docs = [4751, 4624, 4407, 4538, 4828, 4444, 4702, 4896, 4576, 4483]
for i in v1_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v1"].append(fname)
for i in v2_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v2"].append(fname)
for i in v3_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v3"].append(fname)
for i in v4_docs:
    fname = "./test_files/article_" + str(i) + ".html"
    test_files["v4"].append(fname)
```

```python
def clean_text(article_path):
    clean = ""
    with open(article_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, 'lxml')
        # get article title and text
        title = soup.title.text
        body_tags = soup.select("div#body")
        for body in body_tags:
            text = body.get_text()
            # remove footnote markers
            text = re.sub(r'\.\d+', '.', text)
            text = re.sub(r'\.”\d+', '.”', text)
            # remove unnecessary UI text and new lines
            text = re.sub(r'Click for larger view', '', text, flags=re.IGNORECASE)
            text = re.sub(r'View full resolution', '', text, flags=re.IGNORECASE)
            text = re.sub(r'\[End Page \d+\]', '', text, flags = re.IGNORECASE)
            text = re.sub(r'\n+', '\n', text)
            # remove image credit lines
            text = re.sub(r'^.*USHMM.*$', '', text, flags=re.MULTILINE)
            # add cleaned body text to string that will be returned
            clean += text
            break
    print(clean)
```

```python
for vol, articles in test_files.items():
    for article_path in articles:
        clean_text(article_path)
```
