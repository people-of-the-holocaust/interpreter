```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
```

```python
# set up base url for article pages and for index search
base_article_url = "https://muse.jhu.edu"
volume_urls = [
    "https://muse.jhu.edu/resource_group/13",
    "https://muse.jhu.edu/resource_group/59",
    "https://muse.jhu.edu/resource_group/91",
    "https://muse.jhu.edu/resource_group/111"]
# create session to save cookies
session = requests.Session()
session.get(base_article_url)
```

```python
# helper function - get unique 10 random numbers
def get_random_nums(num_articles):
    nums = set()
    while len(nums) != 10:
        # randint: inclusive start and inclusive end
        n = random.randint(0, num_articles-1)
        nums.add(n)
    print(nums)
    return list(nums)
```

```python
# helper function - get all url endings for all content-articles in the volume given the soup
def get_article_links(v_soup):
    # list of toc, separated by ecg sections
    toc_sections = v_soup.find_all(class_="toc_subgroup")
    # list of article links
    links = []
    for section in toc_sections:
        # select img tags with alt text "HTML icon" and has an a parent tag
        html_icons = section.select("a > img[alt*='HTML icon']")
        # save the hrefs of the parent a tags into a list
        hrefs = [img.parent.get('href') for img in html_icons]
        # add to large list of links
        links += hrefs
    print(len(links))
    return links
```

```python
# helper function - save article content to file
def save_article_content(body, url):
    # url format: /document/id
    article_id = url.split("/")[2]
    file_name = "test_files/article_" + article_id + ".txt"
    with open(file_name, 'w') as f:
        f.write(body)
```

```python
# helper function - get the content of all randomly selected articles given the url ending
def get_article_content(links, nums):
    for n in nums:
        # request html for n article
        article_response = requests.get(base_article_url+links[n])
        article_response.raise_for_status()
        article_soup = BeautifulSoup(article_response.text, 'lxml')
        print(article_soup)
        article_body = article_soup.find_all(id="body")
        # if body exists in html, convert to text, use helper function to save to file
        if article_body:
            content = article_body.get_text(strip=True)
            save_article_content(content, links[n])
```

```python
for vol in volume_urls:
    print("VOLUME:", vol)
    # get content of the volume page
    v_response = requests.get(vol)
    v_soup = BeautifulSoup(v_response.text, 'lxml')
    links = get_article_links(v_soup)
    nums = get_random_nums(len(links))
    get_article_content(links, nums)
```
