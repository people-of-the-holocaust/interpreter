```python
import requests
from transformers import pipeline
from bs4 import BeautifulSoup
```

```python
#loading from test data
test_file = open("dachau_test.txt", 'r')
raw_text = test_file.read()
```

```python
# removing HTML tags for NER compatibility
text = BeautifulSoup(raw_text, "html.parser").get_text(separator=" ", strip=True)
```

```python
ner = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
print(f"\n=== Entities found in dachau_test ===")
entities = ner(text[:5000])  # you can increase this limit if your text isn’t huge
for e in entities:
    print(f"{e['word']} ({e['entity_group']}, score={e['score']:.2f})")
```
