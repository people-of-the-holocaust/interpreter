```python
import pandas as pd
```

```python
test_files = ["./test_files/auflag5_test.txt", "./test_files/braunschweig_test.txt",
              "./test_files/dachau_test.txt", "./test_files/dulagcorinth_test.txt",
              "./test_files/warsaw_test.txt"]
```

```python
# load in people table from scraping process
df = pd.read_csv('../ecg_interpreter/people_table.csv')
df = df.drop('Unnamed: 0', axis=1)
# save names in a list
people_names = df['Name'].tolist()
```

```python
# find all instances of a person's name, save the result in a dict
def find_names(article):
    key_sentences = {}
    for name in people_names:
        if name in article:
#             print("FOUND", name)
            name_index = article.find(name)
            end_of_sent = article[name_index:].find('.') + 1 + name_index
            start_of_sent = article[:name_index].rfind('.') + 1
#             print(article[start_of_sent:end_of_sent])
            key_sentences[name] = article[start_of_sent:end_of_sent]
    return key_sentences
```

```python
def pretty_print(dct):
    print("NEW ARTICLE")
    for key, item in dct.items():
        print("NAME:", key)
        print("SENTENCE:", item)
    print()
```

```python
for article_path in test_files:
    with open(article_path, 'r') as f:
        article_content = f.read()
    key_sentences = find_names(article_content)
    pretty_print(key_sentences)
```
